```python
circle1 = Circle(x=0, y=0, radius=1) # unit circle
circle2 = Circle(x=1, y=1, radius=1)
rectangle = Rectangle(x=0, y=0, width=1, height=1)
```

```python
print(circle1 == circle2) # True
print(circle2 == rectangle) 
```
